- `torchvision.transforms` module in PyTorch provides a variety of image transformation operations, which are often used for preprocessing images before feeding them into a neural network. Here an Example of how you might use it.

```python
import torchvision.transforms as transforms
from PIL import Image 
import torch
```

## step 1. Load the image

```python
img = Image.open('datasets/exmaple.jpeg')
img
```

## step 2. Transform the image

```python
transform = transforms.Compose([
    transforms.Resize((254, 256)),  # Resize the image to 256x256 pixels
    transforms.ToTensor(),          # convert image to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406],  # These specific values are typically the mean pixel values for the red, green, and blue channels, respectively, calculated from a large dataset of images, such as the ImageNet dataset.
                         
                        # Red channel: Subtract 0.485 and divide by 0.229
                        # Green channel: Subtract 0.456 and divide by 0.224
                        # Blue channel: Subtract 0.406 and divide by 0.225

                         std=[0.229, 0.224, 0.225])    
])


# Apply transformations 
transformed_img = transform(img)

# Print the shape of the transformed image tensor 
print(transformed_img.shape)   # output: torch.size([3, 256, 256])
```

```python
img = Image.open('datasets/example.jpeg')
transformed_img = transform(img)
transformed_img
```

## step 3. Reverse the Transform image

```python
reverse_transform = transforms.Compose([
    transforms.Normalize(
        mean=[-0.485 / 0.229, -0.456 / 0.224, -0.406 / 0.225],
        std = [1/0.229, 1/0.224, 1/0.255]
    ),
    transforms.ToPILImage()
])


# Apply Reverse Transform and Display Image 
reversed_img = reverse_transform(transformed_img)
reversed_img.show()
```
